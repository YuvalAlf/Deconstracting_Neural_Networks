# Prediction the Function F(x1,x2,x3,x4)=(x1+x2)^2 + (x3+x4)^2

```python
import numpy as np
import seaborn as sns
import pandas as pd
import tensorflow as tf
import matplotlib.pyplot as plt
from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error, r2_score
from sklearn import tree
from keras.models import Sequential
from tensorflow import keras
from tensorflow.keras import layers
from sklearn.neighbors import KNeighborsRegressor
from keras.utils import plot_model
from keras.models import Model
from keras.layers import Input
from keras.layers import Dense
from keras.callbacks import ModelCheckpoint
from keras.layers.merge import concatenate
from keras.callbacks import EarlyStopping
from keras.models import load_model

%matplotlib inline
%config InlineBackend.figure_format = 'svg'

sns.set()
```

```python
np.random.seed(seed=100)
SAMPLES = 1000
EPOCHS = 1000
```

```python
def F(x1, x2, x3, x4):
    return (x1 + x2) ** 2.0 + (x3 + x4)**2.0
```

## Data, Range = [0,2]

```python
data = pd.DataFrame(np.random.rand(SAMPLES,4) * 2.0, columns=['x1', 'x2', 'x3', 'x4'])
data['F'] = F(data['x1'],data['x2'],data['x3'],data['x4'])
X = data.drop(['F'], axis = 1)
Y = data['F']
data.head(10)
```

```python
_ = sns.distplot(data['F'], kde = False, bins = 50)
```

```python
# Display training progress by printing a single dot for each completed epoch
class PrintDot(keras.callbacks.Callback):
  def on_epoch_end(self, epoch, logs):
    if epoch % 100 == 0: print('')
    print('.', end='')

def train_model(model, X, Y):
    history = model.fit(X, Y, epochs=EPOCHS, verbose=0, callbacks=[PrintDot()])
    return history, model
```

```python
fully_nn_model = Sequential()
fully_nn_model.add(Dense(6, activation='relu', input_dim=4))
#fully_nn_model.add(Dense(6, activation='relu'))
fully_nn_model.add(Dense(1))
   
fully_nn_model.compile(loss='mean_squared_error', optimizer='adam')
```

```python
history, fully_nn_model = train_model(fully_nn_model, X, Y)
```

```python
from keras import backend as K

def getLastLayerOutput(input_):
    input1 = fully_nn_model.input               # input placeholder
    output1 = [layer.output for layer in fully_nn_model.layers]# all layer outputs
    fun = K.function([input1, K.learning_phase()],output1)# evaluation function
    t = np.array(input_)[np.newaxis]
    layer_outputs = fun([t, 1])
    return layer_outputs[-2]

def getDerives(input1):
    eps = 0.0001
    output = getLastLayerOutput(input1)
    derives = []
    for index in range(len(input1)):
        newInput = input1[:]
        newInput[index] += eps
        output_ = getLastLayerOutput(newInput)
        d = (output_ - output) / eps
        derives.append(d.squeeze())
    return derives
getDerives([0,2,0,0.5])
```

```python
pointsRange = np.linspace(0,1,4)
samplePoints = ([x1,x2,x3,x4] for x1 in pointsRange for x2 in pointsRange for x3 in pointsRange for x4 in pointsRange)
means = [np.zeros(6) for i in range(4)]
amount = 0
for point in samplePoints:
    amount += 1
    for index, derive in enumerate(getDerives(point)):
        means[index] += derive
for i in range(len(means)):
    means[i] = means[i] / amount
means
```

```python
def show_history(history):
    history_df = pd.DataFrame(history.history)
    history_df['epoch'] = history.epoch
    history_df['log_loss_1'] = np.log(history_df['loss'] + 1)
    sns.lineplot(data = history_df, y='log_loss_1', x='epoch')
    plt.gca().set_ylim(bottom=0)
```

```python
def show_predictions(model, X, Y):
    predicted = model.predict(X).flatten()
    expected = Y
    print(((Y-predicted)**2.0).mean())
    plt.scatter(expected, predicted, alpha=.2, color='yellowgreen')
    plt.xlabel('Real Values [F]')
    plt.ylabel('Predictions [F]')
    plt.axis('equal')
    plt.axis('square')
    plt.xlim([0,plt.xlim()[1]])
    plt.ylim([0,plt.ylim()[1]])
    _ = plt.plot([0, 100], [0, 100])
```

```python
def build_halved_model():
    input1 = Input(shape=(2,))
    hidden1_1 = Dense(4, activation='relu')(input1)
    out1 = Dense(1)(hidden1_1)
    
    input2 = Input(shape=(2,))
    hidden2_1 = Dense(4, activation='relu')(input2)
    out2 = Dense(1)(hidden2_1)
    
    merge = concatenate([out1, out2])
    hidden = Dense(4, activation='relu')(merge)
    #hidden2 = Dense(4, activation='relu')(hidden1)
    output = Dense(1)(hidden)
    
    model = Model(inputs=[input1, input2], outputs=output)
    
    model.compile(loss='mean_squared_error', optimizer='adam')
    
    return model
```

```python
good_halved_model = build_halved_model()
goodX = [X[['x1', 'x2']], X[['x3', 'x4']]]
good_history, good_halved_model = train_model(good_halved_model, goodX, Y)
```
